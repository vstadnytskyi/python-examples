## Working with images obtain from machine vision cameras
A typical machine vision cameras such as visiable cameras or depth cameras usually return data as 2D array for depth or monochromatic cameras, or 3D array for colored camera. The dimensions are (height, width, color) or (rows, colomns, color).

As an example, we can get images from image processing library skimage and pretend we got that data from a different cameras.

```python
colored_image = skimage.data.cat()
monochrom_image = skimage.data.coins()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure()
plt.imshow(colored_image)
plt.title('colored image')
```

```python
plt.figure()
plt.imshow(monochrom_image)
plt.title('monochromatic image')
```
